# Observations on the type of model

1. More number of dense layers - longer time
2. More neurons in dense layers (2000+ 2 layers) - Stagnation in loss around 80 - too much
3. Smaller starting conv layer - faster training with lesser loss
4. For dense layers, best to have two/three layers with decreasing number of neurons... Any more, and the speed would be affected. However, if ready to compromise with training time, more dense layers give out better results.
5. Increasing the number of Conv layers beyond a point doesn't help. The model becomes both slow in training and in converging to a lower level of loss.

Tests with jump5 model - 
1. Smaller and larger dense layers - speed is affected but I believe the model learns more with large initial dense layers.
